```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Função para desenhar o grafo com um título
def desenhar_molecula(grafo, posicoes, titulo):
    plt.figure(figsize=(4, 4))
    nx.draw(grafo, pos=posicoes, with_labels=True, node_size=1200, node_color='skyblue', font_weight='bold')
    plt.title(titulo)
    plt.axis('off')
    plt.show()
```

```python
# ------------------------------
# Molecula de água (H2O)
# ------------------------------

# Criamos um grafo para a molécula de água
agua = nx.Graph()

# Adicionamos os átomos como nós
agua.add_node("O")      # Oxigênio
agua.add_node("H1")     # Hidrogênio 1
agua.add_node("H2")     # Hidrogênio 2

# Adicionamos as ligações covalentes como arestas
agua.add_edge("O", "H1")
agua.add_edge("O", "H2")

# Posição manual dos átomos para visualização clara
pos_agua = {
    "O": (0, 0),
    "H1": (-1, -1),
    "H2": (1, -1)
}

# Visualização
desenhar_molecula(agua, pos_agua, "Molécula de Água (H₂O)")
```

```python

# ------------------------------
# Molécula de dióxido de carbono (CO2)
# ------------------------------

# Criamos um grafo para a molécula de CO2
co2 = nx.Graph()

# Adicionamos os átomos como nós
co2.add_node("C")      # Carbono
co2.add_node("O1")     # Oxigênio 1
co2.add_node("O2")     # Oxigênio 2

# Adicionamos as ligações covalentes como arestas
co2.add_edge("C", "O1")
co2.add_edge("C", "O2")

# Posição dos átomos
pos_co2 = {
    "C": (0, 0),
    "O1": (-1, 0),
    "O2": (1, 0)
}

# Visualização
desenhar_molecula(co2, pos_co2, "Molécula de Dióxido de Carbono (CO₂)")
```
